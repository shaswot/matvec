```python
# generate hardware error profile
import numpy as np
import random
import os
```

```python
seed = 20210907
print("Seed: ", seed)
np.random.seed(seed)
random.seed(seed)
os.environ['PYTHONHASHSEED']=str(seed)
```

```python
NO_OF_THREADS = 64
NO_OF_CORES = 64
```

```python
error_profile = np.random.uniform(low=0.0, high=0.7, size = (NO_OF_CORES, NO_OF_THREADS))
```

```python
error_profile
```

```python
import matplotlib.pyplot as plt
plt.imshow(error_profile, 
           cmap='hot', 
           interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.imshow(error_profile[0:2,:], 
           cmap='hot', 
           interpolation='nearest')
plt.tight_layout()
plt.show()
```
